## Mission 3

이번 미션에서는 Figure S2A 아랫쪽에 표시되는 clustered error 분석 (CIMS; Crosslinking-induced mutation sites)을 합니다. 자외선에 의해서 단백질이 붙어있다가 단백질 분해효소로 단백질을 뜯어내고 남은 자리에서는 reverse-transcriptase가 오류를 자주 냅니다. 오류율을 분석해서 단백질이 앉아 있었던 자리를 분석해 봅시다.

````
import pysam
import subprocess
import pandas as pd
import numpy as np
import re # Import re for potential future use or more robust parsing, though current fix is simpler.

# ==========================================
# 1. 설정 및 데이터 세팅 (let-7 패밀리 비교군 확장)
# ==========================================
fasta_path = "mm9.fasta"
window = 25  # 앞뒤 25bp (총 50bp 서열 분석)

# 분석 대상 target (유전자명, 염색체, CIMS 최고 엔트로피 피크 위치)
# 학기말 텀프로젝트용 빈틈없는 분석을 위해 비교군(let7f-1)을 함께 분석합니다.
targets = {
    "Mirlet7g": {"chrom": "chr9", "peak_pos": 106056062},   # 106056000 ~ 106056125 사이의 Peak
    "Mirlet7f-1": {"chrom": "chr13", "peak_pos": 48691350}  # 48691300 ~ 48691400 사이의 Peak
}

def get_rna_structure(seq):
    """ViennaRNA를 호출하여 구조와 MFE를 반환하는 함수"""
    result = subprocess.run(
        ['RNAfold'],
        input=seq.encode('utf-8'),
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE
    )
    output = result.stdout.decode('utf-8').strip().split('\n')
    if result.returncode == 0 and len(output) > 1:
        structure_mfe_parts = output[1].split()
        if len(structure_mfe_parts) >= 2:
            structure = structure_mfe_parts[0]
            mfe_str = structure_mfe_parts[1].strip('()')
            try:
                mfe = float(mfe_str)
            except ValueError:
                # If MFE string is empty or cannot be converted, assign 0.0
                # This handles cases where RNAfold might output '()' for MFE.
                if mfe_str.strip() == '':
                    mfe = 0.0
                else:
                    raise RuntimeError(f"RNAfold output parsing error: Could not convert '{mfe_str}' to float from line '{output[1]}'")
            return structure, mfe
        else:
            raise RuntimeError(f"RNAfold output parsing error: Unexpected output format in line '{output[1]}'")
    else:
        # Include stderr in the error message for better debugging
        error_message = result.stderr.decode('utf-8') or 'No specific error from RNAfold'
        raise RuntimeError(f"RNAfold 에러: {error_message}. Full stdout: {result.stdout.decode('utf-8')}")

def analyze_structural_context(dot_bracket, peak_index, context_window=3):
    """CIMS 피크 좌표 주변의 구체적인 구조적 형태(Stem vs Loop)를 판정하는 함수"""
    # 피크 중심 주변부 추출
    start = max(0, peak_index - context_window)
    end = min(len(dot_bracket), peak_index + context_window + 1)
    local_structure = dot_bracket[start:end]

    # 구조 카운트
    stem_count = local_structure.count('(') + local_structure.count(')')
    loop_count = local_structure.count('.')

    # 대표 구조 판정
    if stem_count > loop_count:
        biotype = "Stem (Double-stranded)"
    else:
        biotype = "Loop/Bulge (Single-stranded)"

    return biotype, local_structure

# ==========================================
# 2. 메인 분석 파이프라인 가동
# ==========================================
fasta = pysam.FastaFile(fasta_path)
results_list = []

print("## [분석 시작] RNA 2차 구조 및 CIMS 위치 매핑\n")

for name, info in targets.items():
    # 2-1. 서열 추출
    start_pos = info["peak_pos"] - window
    end_pos = info["peak_pos"] + window
    seq = fasta.fetch(info["chrom"], start_pos, end_pos).upper()

    # 2-2. RNAfold 구조 예측
    dot_bracket, mfe = get_rna_structure(seq)

    # 2-3. CIMS 피크의 상대적 위치 계산 (70bp 서열 중 딱 중앙인 35번째 인덱스)
    peak_idx_in_seq = window
    peak_base = seq[peak_idx_in_seq]
    peak_struct = dot_bracket[peak_idx_in_seq]

    # 2-4. 피크 주변부(±3bp) 미세 구조 환경 분석
    struct_type, local_str = analyze_structural_context(dot_bracket, peak_idx_in_seq, context_window=3)

    # 결과 저장
    results_list.append({
        "Gene": name,
        "MFE (kcal/mol)": mfe,
        "Peak_Base": peak_base,
        "Peak_Structure": peak_struct,
        "Local_Context(±3bp)": local_str,
        "Structural_Preference": struct_type
    })

# ==========================================
# 3. 데이터프레임 변환 및 가상 돌연변이(In silico Mutagenesis) 시뮬레이션
# ==========================================
df = pd.DataFrame(results_list)

print("### 1) 유전자별 구조 및 열역학적 안정성 비교")
print(df.to_markdown(index=False))
print("\n" + "-"*40 + "\n")

print("### 2) 인위적 돌연변이(Mutation)에 따른 구조 붕괴 시뮬레이션 (Mirlet7g 대상)")
# Mirlet7g의 피크 위치 염기를 강제로 바꾸어 구조가 얼마나 변하는지 테스트
mirlet7g_seq = fasta.fetch(targets["Mirlet7g"]["chrom"], targets["Mirlet7g"]["peak_pos"] - window, targets["Mirlet7g"]["peak_pos"] + window).upper()

# 원래 중앙 염기 확인 후, 완전히 결합을 방해할 다른 염기로 치환 (예: A -> C 또는 G -> T)
original_base = mirlet7g_seq[window]
mutated_base = "C" if original_base != "C" else "A"

mutated_seq = list(mirlet7g_seq)
mutated_seq[window] = mutated_base
mutated_seq = "".join(mutated_seq)

# 돌연변이 서열 구조 예측
mut_dot_bracket, mut_mfe = get_rna_structure(mutated_seq)
orig_dot_bracket, orig_mfe = get_rna_structure(mirlet7g_seq)

print(f"• 야생형(Wild-Type) 서열 내 피크 염기: {original_base} (MFE: {orig_mfe} kcal/mol)")
print(f"• 돌연변이(Mutant) 서열 내 피크 염기   : {mutated_base} (MFE: {mut_mfe} kcal/mol)")
print(f"• 야생형 구조: {orig_dot_bracket}")
print(f"• 변이형 구조: {mut_dot_bracket}")

mfe_diff = mut_mfe - orig_mfe
print(f"• 에너지 변화량 (ΔΔG): {mfe_diff:+.2f} kcal/mol")
if mfe_diff > 1.5:
    print("=> 결론: 돌연변이 도입 시 RNA 구조가 열역학적으로 크게 불안정해지며, 기존 RBP 결합 구조가 파괴될 가능성이 높음.")
else:
    print("=> 결론: 돌연변이 도입 후에도 전체적인 구조적 안정성은 크게 변하지 않음.")
````
